# Non-Personalized Recommendation

One of the most basic ways to make recommendations is to go with the knowledge of the crowd and recommend what is already the most popular.

We will calculate how often each book in the dataset has been read and find the most frequently read books.

Just because a book has been read by a lot of people doesn't necessarily mean readers enjoyed it.
Therefore, we shall compute for the average ratings of the most popular books and recommend them. 

```python
# Create a list of only books appearing > 100 times in the dataset
book_popularity = df["Title"].value_counts()
popular_books = book_popularity[book_popularity > 100].index

# Use this list of popular books to filter the original DataFrame
popular_books_ratings =  df[df["Title"].isin(popular_books)]

# Find the average rating given to these popular books
popular_books_average_ratings = popular_books_ratings[["Title", "Rating"]].groupby('Title').mean()

popular_books_average_ratings.sort_values(by="Rating", ascending=False).head(10)
```

```python
new_df = pd.DataFrame(df.groupby("Title").mean()["Rating"])
new_df['Num_of_ratings'] = pd.DataFrame(df.groupby('Title').count()['Rating'])

new_df.head()
```

```python
# Visualizing top 10 books with highest no. of ratings
new_df.sort_values(by=['Num_of_ratings','Rating'],ascending=False).head(10)
```

### Finding Pairs

This is an attempt to find all pairs of movies or all permutations of pairs of books that have been read by the same person. And make recommendations on the assumption that similar people read similar content.

```python
from itertools import permutations

# Creating a function to find all permutations
def find_book_pairs(x):
  pairs = pd.DataFrame(list(permutations(x.values, 2)),
                       columns=['book_a', 'book_b'])
  return pairs

# Applying the function to the title column and resetting the index
book_pairs = reduced_df.groupby('User_Id')['Title'].apply(
  find_book_pairs).reset_index(drop=True)

print(book_pairs)
```

Now, We have a clean table of all of the books that were read by the same user, which can be used to find the most commonly paired books.

### Counting Pairs

```python
# Calculating how often each item in movie_a occurs with the items in movie_b
book_pairs_counts = book_pairs.groupby(['book_a', 'book_b']).size()

# Converting the results to a DataFrame and resetting the index
book_pairs_df = book_pairs_counts.to_frame(name='size').reset_index()

# Sorting the counts from highest to lowest
book_pairs_df = book_pairs_df.sort_values('size', ascending=False)

book_pairs_df.head(10)
```

# Making Recommendation
